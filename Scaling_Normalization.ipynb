# Scaling vs Normalization
Scaling = Your are changing the range of your data
Normalization = You are changing the shape of the distribution of your data

### Scaling
You transform data within a specific scale, like 0-100 or 0-1. Is used for example for: 
    1. Support vector machines (SVM)
    K-nearest neighbors (KNN)

Help compare different variables on equal footing. FOr example, if you wanna compare weight with height you must Scale your data in order to compare these two variables. 

```python
import pandas as pd
import numpy as np

from scipy import stats

from mlxtend.preprocessing import minmax_scaling

import seaborn as sns   
import matplotlib.pyplot as plt
```

## Example to see how scaling works

```python
#generating 1000 data points randomly drawn from exponential distribution
original_data = np.random.exponential(size=1000)

#mix-max scale the data between 0 and 1
scaled_data = minmax_scaling(original_data, columns=[0])

#plot both together to compare
fig, ax = plt.subplots(1, 2, figsize=(15, 3))
sns.histplot(original_data, ax=ax[0], kde=True, legend=False)
ax[0].set_title("Original Data")
sns.histplot(scaled_data, ax=ax[1], kde=True, legend=False)
ax[1].set_title("Scaled data")
plt.show()
```

We keep the same shape, but instead of going from 0 to 6, we have changed the range 0-1

### Example of Normalization
Normal distribution: Also know as the "bell curve", this is a specific statistical distribution where a roughly equal observations fall above and below the mean, the mean and the median are the same, and there more observations closer to the mean. The normal distribution is also know as the Gaussian Distribution. 
You are going to normlaize your data when using machine learning or statistics technique that assumes your data is normally distributed like: 
1. Linear discriminant analysis (LDA)
2. Gaussian naive Bayes

```python
#normalize the exponential data with box-cox transformation
normalized_data = stats.boxcox(original_data)

#plot both together to compare
fig, ax = plt.subplots(1, 2, figsize=(15, 3))
sns.histplot(original_data, ax=ax[0], kde=True, legend=False)
ax[0].set_title("Original Data")
sns.histplot(normalized_data, ax=ax[1], kde=True, legend=False)
ax[1].set_title("Normalized data")
plt.show()
```
